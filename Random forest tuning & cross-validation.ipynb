```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score,\
f1_score, confusion_matrix, ConfusionMatrixDisplay

from sklearn.ensemble import RandomForestClassifier
import pickle
```

```python
file = 'Churn_Modelling.csv'
df_original = pd.read_csv(file)
df_original.head()
```

```python
# Feature engineering
churn_df = df_original.drop(['RowNumber', 'CustomerId', 'Surname', 'Gender'], axis=1)
churn_df.head()
```

```python
#Feature transformation
churn_df2 = pd.get_dummies(churn_df, drop_first='True')
churn_df2.head()
```

```python
#Split the data
y = churn_df2["Exited"]
X = churn_df2.copy()
X = X.drop("Exited", axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=y, random_state=42)
```

```python
# The classifier (rf)
# The dictionary of hyperparameters to search over (cv_params)
# The set of scoring metrics (scoring)
# The number of cross-validation folds you want (cv=5)
# The scoring metric that you want GridSearch to use when it selects the "best" model
#(i.e., the model that performs best on average over all validation folds) (refit='f1')
# Fit the data (X_train, y_train) to the GridSearchCV object (rf_cv).

rf = RandomForestClassifier(random_state=0)

cv_params = {'max_depth': [2,3,4,5, None], 
             'min_samples_leaf': [1,2,3],
             'min_samples_split': [2,3,4],
             'max_features': [2,3,4],
             'n_estimators': [75, 100, 125, 150]
             }  

scoring = {'accuracy', 'precision', 'recall', 'f1'}

rf_cv = GridSearchCV(rf, cv_params, scoring=scoring, cv=5, refit='f1')

rf_cv.fit(X_train, y_train)
```

```python
#Random forest validation on separate datase
#Using pickle to save a fit model
path = '/home/jovyan/work/
```

```python
# Pickle the model
with open(path+'rf_cv_model.pickle', 'wb') as to_write:
    pickle.dump(rf_cv, to_write)
```
